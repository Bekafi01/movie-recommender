```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv("..\\data\\ml - 1m\\ratings.csv")
print(ratings.head())
print(ratings.info())
```

```python
movies = pd.read_csv("..\\data\\ml - 1m\\movies.csv")
print(movies.head())
print(movies.info())
```

```python
imdb = pd.read_csv("..\\data\\movie_metadata.csv")
print(imdb.info())
```

```python
# Missing values in ratings
missing_ratings = ratings.isnull().sum()
print("Missing values in ratings dataset: ")
print(missing_ratings)
```

```python
# Missing values in movies
missing_movies = movies.isnull().sum()
print("Missing values in movies dataset: ")
print(missing_movies)
```

```python
# Missing values in imdb
missing_imdb = imdb.isnull().sum()
print("Missing values in imdb dataset: ")
print(missing_imdb)
```

```python
# Range of ratings
plt.figure(figsize=(8, 5))
ratings['rating'].hist(bins=10, edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks(range(1, 6))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Number of unique users and movies
num_users = ratings['userId'].nunique()
num_movies = ratings['movieId'].nunique()
print(f"Number of unique users: {num_users}")
print(f"Number of unique movies: {num_movies}")
```
